```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("housing.csv")
```

```python
df
```

```python
#stratifiedSuffleSplit

df["income_cat"]=pd.cut(df["median_income"],bins=[ 0 , 1.5 , 3 , 4.5 , 6 , np.inf ],labels=[1,2,3,4,5])
```

```python
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split=StratifiedShuffleSplit(n_splits=1,test_size=.2,random_state=42)

for train_index,test_index in split.split(df,df["income_cat"]):
    strat_train_set=df.loc[train_index]
    strat_test_set=df.loc[test_index]
```

```python
strat_train_set
```

```python
strat_test_set
```

```python
for i in (strat_train_set,strat_test_set):
    i.drop("income_cat",inplace=True,axis=1)
```

```python
strat_train_set
```

```python
strat_test_set
```

```python
housing=strat_train_set.drop("median_house_value",axis=1)
housing_label=strat_train_set["median_house_value"].copy()
```

```python
housing
```

```python
# HANDLING_MISSING_VALUES USING IMPUTER
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy="mean")
housing_num=housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Sample DataFrame (replace this with your actual `housing` DataFrame)
housing = pd.DataFrame({
    'ocean_proximity': ['NEAR BAY', 'INLAND', 'NEAR OCEAN', 'NEAR BAY', 'ISLAND']
})

# Select the categorical columns (replace with your actual categorical subset)
housing_cat = housing[['ocean_proximity']]  # This remains a DataFrame

# OneHotEncoder setup
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform
housing_encoded = encoder.fit_transform(housing_cat)

# Convert to DataFrame and align index with `housing_cat`
housing_encoded_df = pd.DataFrame(
    housing_encoded,
    columns=encoder.get_feature_names_out(housing_cat.columns),
    index=housing_cat.index  # ✅ works because `housing_cat` is a DataFrame
)


```
