```python
#here we have created a new column based on the value of median_income as median income is a strong predictor of house prices
df['income_cat']=pd.cut(df['median_income'],bins=[0, 1.5 , 3 , 4.5 , 6 , np.inf ],labels=[1,2,3,4,5])
```

```python
df
```

```python
#We can plot these income categories to visualize the distribution:
import matplotlib.pyplot as plt
df["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Income Categories Distribution") 
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
# here we are split the test and train set where test has 20% of the whole data
split=StratifiedShuffleSplit(n_splits=1,test_size=.2,random_state=42)

for train_index,test_index in split.split(df,df['income_cat']):
      strat_train_set=df.loc[train_index]
      strat_test_set=df.loc[test_index]
```

```python
strat_train_set["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Income Categories Distribution for train set") 
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```

```python
strat_test_set["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Income Categories Distribution for test set") 
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```
